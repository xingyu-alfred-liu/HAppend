## Guide
In this tutorial, you can learn how to load all data into one dictionary, plot bandstructure and absorption from PAH101 dataset.

```python
# Load packages
from pah101tools.plot import PlotAgent
```

### Load dataset
To gather all data, you will need to specify the path point to the data folder. After plot_agent is setup, you can access all data through `plot_agent.data_dict`

```python
root_path = "/Users/alfred/Desktop/work/PAH101/PAH101Plot/data"
plot_agent = PlotAgent(root_path=root_path)

print(plot_agent.data_dict.keys())
```

### Plot Bandstructure
Pass in the `struct_id`, and `savefig_path` to specify which material to plot bandstructure for, and where to save the plot. If `savefig_path` is not specified, the plot will NOT be saved. Here we use "ABECAL" and "band.png" as an example. Similar for absorption.

```python
struct_id = "ABECAL"
savefig_path = "band.png"

bandstruct_fig = plot_agent.plot_bandstructure(
    struct_id=struct_id, 
    savefig_path=savefig_path
)

bandstruct_fig.show()
```

### Plot Absorption

```python
struct_id = "ABECAL"
savefig_path = "absorp.png"

absorption_fig = plot_agent.plot_absorption(
    struct_id=struct_id,
    savefig_path=savefig_path
)

absorption_fig.show()
```

```python
import json
for k in plot_agent.data_dict:
    if k == "ABECAL": continue
    plot_agent.data_dict[k]["gwbse"]["absorption"]["solar"] = plot_agent.data_dict["ABECAL"]["gwbse"]["absorption"]["solar"]
    
    with open(k+".json", "w") as f:
        json.dump(plot_agent.data_dict[k], f, indent=4)
```
